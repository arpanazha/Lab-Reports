```python
# ============================================================
# Lab Session 2 - Part I: Working with Dictionary
# ============================================================

# --- Initial Student Dictionary ---
student = {
    "name": "Ram",
    "age": 20,
    "faculty": "BCE"
}

# ----------------------------------------
# 1. Create Dictionary
# ----------------------------------------
d = {"a": 1, "b": 2}
print("Created Dictionary:", d)

# ----------------------------------------
# 2. Access Value using key
# ----------------------------------------
print("\nAccess Value d['a']:", d["a"])

# ----------------------------------------
# 3. Access Value using get()
# ----------------------------------------
print("Access using get():", d.get("a"))

# ----------------------------------------
# 4. Add Item
# ----------------------------------------
d["c"] = 3
print("\nAfter Adding 'c':", d)

# ----------------------------------------
# 5. Update Value
# ----------------------------------------
d["a"] = 10
print("After Updating 'a' to 10:", d)

# ----------------------------------------
# 6. Remove Item using del
# ----------------------------------------
del d["b"]
print("\nAfter Deleting 'b':", d)

# ----------------------------------------
# 7. Remove Item using pop()
# ----------------------------------------
# First add 'b' back so we can pop it
d["b"] = 5
popped_value = d.pop("b")
print("Popped value:", popped_value)
print("After pop('b'):", d)

# ----------------------------------------
# 8. Remove Last Item using popitem()
# ----------------------------------------
# Add some items first
d["x"] = 100
d["y"] = 200
last_item = d.popitem()
print("\nLast item removed by popitem():", last_item)
print("After popitem():", d)

# ----------------------------------------
# 9. Copy Dictionary
# ----------------------------------------
new_d = d.copy()
print("\nCopied Dictionary:", new_d)

# ----------------------------------------
# 10. Get All Keys
# ----------------------------------------
print("\nAll Keys:", d.keys())

# ----------------------------------------
# 11. Get All Values
# ----------------------------------------
print("All Values:", d.values())

# ----------------------------------------
# 12. Get All Items
# ----------------------------------------
print("All Items:", d.items())

# ----------------------------------------
# 13. Check if Key Exists
# ----------------------------------------
print("\n'a' in d:", "a" in d)
print("'z' in d:", "z" in d)

# ----------------------------------------
# 14. Get Length of Dictionary
# ----------------------------------------
print("\nLength of d:", len(d))

# ----------------------------------------
# 15. Loop Through Keys
# ----------------------------------------
print("\nLoop Through Keys:")
for k in d:
    print("  Key:", k)

# ----------------------------------------
# 16. Loop Through Values
# ----------------------------------------
print("\nLoop Through Values:")
for v in d.values():
    print("  Value:", v)

# ----------------------------------------
# 17. Loop Through Items (Key-Value Pairs)
# ----------------------------------------
print("\nLoop Through Items:")
for k, v in d.items():
    print(f"  Key: {k}, Value: {v}")

# ----------------------------------------
# 18. Clear Dictionary
# ----------------------------------------
d.clear()
print("\nAfter clear():", d)
```

```python
# ============================================================
# Lab Session 2 - Part II: Working with Pandas DataFrame
# ============================================================

import pandas as pd

# ----------------------------------------
# 1. Create a DataFrame
# ----------------------------------------
df = pd.DataFrame({
    "Name"  : ["Ram", "Sita", "Hari", "Gita", "Mohan"],
    "Age"   : [20, 22, 19, 21, 23],
    "Marks" : [85, 78, 92, 74, 88]
})
print("Created DataFrame:")
print(df)

# ----------------------------------------
# 2. Display First 5 Rows
# ----------------------------------------
print("\nFirst 5 Rows (head()):")
print(df.head())

# ----------------------------------------
# 3. Display Last 5 Rows
# ----------------------------------------
print("\nLast 5 Rows (tail()):")
print(df.tail())

# ----------------------------------------
# 4. Total Number of Rows and Columns
# ----------------------------------------
print("\nShape (rows, columns):", df.shape)

# ----------------------------------------
# 5. Display All Column Names
# ----------------------------------------
print("\nColumn Names:", df.columns)

# ----------------------------------------
# 6. Complete Info of DataFrame
# ----------------------------------------
print("\nDataFrame Info:")
df.info()

# ----------------------------------------
# 7. Statistical Summary
# ----------------------------------------
print("\nStatistical Summary:")
print(df.describe())

# ----------------------------------------
# 8. Select Single Column - Name
# ----------------------------------------
print("\nName Column:")
print(df["Name"])

# ----------------------------------------
# 9. Select Multiple Columns - Name and Marks
# ----------------------------------------
print("\nName and Marks Columns:")
print(df[["Name", "Marks"]])

# ----------------------------------------
# 10. First Row using iloc (positional)
# ----------------------------------------
print("\nFirst Row using iloc[0]:")
print(df.iloc[0])

# ----------------------------------------
# 11. First Row using loc (label-based)
# ----------------------------------------
print("\nFirst Row using loc[0]:")
print(df.loc[0])

# ----------------------------------------
# 12. Access Specific Cell Value
# ----------------------------------------
print("\nMarks of First Student using loc[0, 'Marks']:")
print(df.loc[0, "Marks"])

# ----------------------------------------
# 13. Filter Students with Marks > 80
# ----------------------------------------
print("\nStudents with Marks > 80:")
print(df[df["Marks"] > 80])

# ----------------------------------------
# 14. Filter by Age > 20 AND Marks > 75
# ----------------------------------------
print("\nStudents with Age > 20 and Marks > 75:")
print(df[(df["Age"] > 20) & (df["Marks"] > 75)])

# ----------------------------------------
# 15. Add a New Column - Grade
# ----------------------------------------
df["Grade"] = ["A", "B", "A+", "B+", "A"]
print("\nAfter Adding Grade Column:")
print(df)

# ----------------------------------------
# 16. Increase All Marks by 5
# ----------------------------------------
df["Marks"] = df["Marks"] + 5
print("\nAfter Increasing All Marks by 5:")
print(df)

# ----------------------------------------
# 17. Update Marks of Ram to 95
# ----------------------------------------
df.loc[df["Name"] == "Ram", "Marks"] = 95
print("\nAfter Updating Ram's Marks to 95:")
print(df)

# ----------------------------------------
# 18. Remove Age Column
# ----------------------------------------
df_no_age = df.drop("Age", axis=1)
print("\nAfter Removing Age Column:")
print(df_no_age)

# ----------------------------------------
# 19. Remove First Row
# ----------------------------------------
df_no_first_row = df.drop(0, axis=0)
print("\nAfter Removing First Row:")
print(df_no_first_row)

# ----------------------------------------
# 20. Sort by Marks Ascending
# ----------------------------------------
print("\nSorted by Marks (Ascending):")
print(df.sort_values("Marks"))

# ----------------------------------------
# 21. Sort by Marks Descending
# ----------------------------------------
print("\nSorted by Marks (Descending):")
print(df.sort_values("Marks", ascending=False))

# ----------------------------------------
# 22. Concatenate Two DataFrames Vertically
# ----------------------------------------
df1 = pd.DataFrame({
    "Name"  : ["Ram", "Sita"],
    "Age"   : [20, 22],
    "Marks" : [85, 78]
})

df2 = pd.DataFrame({
    "Name"  : ["Hari", "Gita"],
    "Age"   : [19, 21],
    "Marks" : [92, 74]
})

df_combined = pd.concat([df1, df2], ignore_index=True)
print("\nConcatenated DataFrame:")
print(df_combined)

# ----------------------------------------
# 23. Write DataFrame to CSV File
# ----------------------------------------
df.to_csv("students.csv", index=False)
print("\nDataFrame saved to 'students.csv'")

# ----------------------------------------
# 24. Read Data from CSV File
# ----------------------------------------
df_from_csv = pd.read_csv("students.csv")
print("\nDataFrame read from 'students.csv':")
print(df_from_csv)

# ----------------------------------------
# 25. Write DataFrame to Excel File
# ----------------------------------------
df.to_excel("students.xlsx", index=False)
print("\nDataFrame saved to 'students.xlsx'")

# ----------------------------------------
# 26. Read Data from Excel File
# ----------------------------------------
df_from_excel = pd.read_excel("students.xlsx")
print("\nDataFrame read from 'students.xlsx':")
print(df_from_excel)
```

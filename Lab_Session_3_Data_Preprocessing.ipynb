```python
# ============================================================
# Lab Session 3: Data Preprocessing for Machine Learning
# Dataset: Titanic Passenger Dataset
# ============================================================

# ------------------------------------------------------------
# Part A: Loading Dataset and Displaying Basic Information
# ------------------------------------------------------------

# Task 1: Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")

# Task 2: Load the dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("\nDataset loaded successfully!")
print("Shape of dataset:", df.shape)

# Task 3: Display basic information
print("\n----- First 5 Rows -----")
print(df.head())

print("\n----- Dataset Info -----")
df.info()

print("\n----- Statistical Summary -----")
print(df.describe())

# ------------------------------------------------------------
# Part B: Handling Missing Values
# ------------------------------------------------------------

# Task 4: Check missing values
print("\n----- Missing Values Count -----")
print(df.isnull().sum())

# Task 5: Visualize missing values in a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()

# Exercise 2 (part): Average Age BEFORE filling missing values
avg_age_before = df['Age'].mean()
print("\nAverage Age BEFORE filling missing values:", round(avg_age_before, 2))

# Task 6: Fill missing Age values with median
df['Age'].fillna(df['Age'].median(), inplace=True)
print("\nMissing Age values filled with median.")

# Exercise 2 (part): Average Age AFTER filling missing values
avg_age_after = df['Age'].mean()
print("Average Age AFTER filling missing values:", round(avg_age_after, 2))

# Task 7: Fill missing Embarked values with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print("Missing Embarked values filled with mode.")

# Task 8: Drop Cabin column (too many missing values)
df.drop('Cabin', axis=1, inplace=True)
print("Cabin column dropped.")

# Verify missing values after handling
print("\n----- Missing Values After Handling -----")
print(df.isnull().sum())

# ------------------------------------------------------------
# Part C: Encoding Categorical Variables
# ------------------------------------------------------------

# Task 9: Identify categorical attributes
print("\n----- Categorical Columns -----")
print(df.select_dtypes(include='object').columns.tolist())

# Exercise 1: Count male and female passengers (do this BEFORE encoding Sex)
print("\n----- Gender Count (before encoding) -----")
print(df['Sex'].value_counts())

# Task 10: Apply Label Encoding on Sex
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
print("\nLabel Encoding applied on 'Sex' column.")
print("Sex column after Label Encoding (sample):")
print(df['Sex'].head())
print("Mapping: 0 = female, 1 = male (usually)")

# Task 11: Apply One-Hot Encoding on Embarked
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
print("\nOne-Hot Encoding applied on 'Embarked' column.")
print("Columns after One-Hot Encoding:")
print(df.columns.tolist())

# ------------------------------------------------------------
# Part D: Feature Scaling
# ------------------------------------------------------------

# Task 12: Select numerical columns for scaling
num_cols = ['Age', 'Fare']
print("\nNumerical columns selected for scaling:", num_cols)

# Task 13: Apply Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
print("Standardization applied on Age and Fare.")

# Task 14: Observe transformed values
print("\n----- Transformed Age and Fare (first 5 rows) -----")
print(df[['Age', 'Fare']].head())

# ------------------------------------------------------------
# Part E: Outlier Detection
# ------------------------------------------------------------

# Task 15: Draw boxplot for Fare
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare (after Standardization)")
plt.show()

# Task 16: Calculate Inter Quartile Range (IQR)
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("\n----- IQR Calculation for Fare -----")
print("Q1 (25th percentile):", round(Q1, 4))
print("Q3 (75th percentile):", round(Q3, 4))
print("IQR:", round(IQR, 4))
print("Lower Bound:", round(lower, 4))
print("Upper Bound:", round(upper, 4))

# Task 17: Find outliers
outliers = df[(df['Fare'] < lower) | (df['Fare'] > upper)]
print("\nNumber of outliers in Fare:", len(outliers))

# Exercise 3: Detect outliers in the Age column
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_age = Q1_age - 1.5 * IQR_age
upper_age = Q3_age + 1.5 * IQR_age
outliers_age = df[(df['Age'] < lower_age) | (df['Age'] > upper_age)]
print("Number of outliers in Age:", len(outliers_age))

# ------------------------------------------------------------
# Part F: Feature Selection and Final Dataset
# ------------------------------------------------------------

# Task 18: Remove unnecessary columns
df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
print("\nUnnecessary columns dropped: PassengerId, Name, Ticket")
print("Remaining columns:", df.columns.tolist())

# Task 19: Separate features and target
X = df.drop('Survived', axis=1)
y = df['Survived']

print("\nFeatures (X) columns:", X.columns.tolist())
print("Target (y): Survived")

# Task 20: Check final shape
print("\n----- Final Shapes -----")
print("X shape (features):", X.shape)
print("y shape (target):", y.shape)

print("\n----- Final Feature Matrix (first 5 rows) -----")
print(X.head())

print("\n----- Target Variable (first 10 values) -----")
print(y.head(10))

# ============================================================
# Additional Exercises
# ============================================================

print("\n" + "="*60)
print("ADDITIONAL EXERCISES")
print("="*60)

# Exercise 1 already done above (Gender count before encoding)

# Exercise 2 already done above (Average age before & after)

# Exercise 3 already done above (Age outliers)

# Exercise 4: Apply Min-Max Scaling instead of Standardization
print("\n----- Exercise 4: Min-Max Scaling -----")
from sklearn.preprocessing import MinMaxScaler

# Reload original data for clean comparison
df_minmax = pd.read_csv(url)
df_minmax['Age'].fillna(df_minmax['Age'].median(), inplace=True)

minmax_scaler = MinMaxScaler()
df_minmax[['Age', 'Fare']] = minmax_scaler.fit_transform(df_minmax[['Age', 'Fare']])

print("Age and Fare after Min-Max Scaling (first 5 rows):")
print(df_minmax[['Age', 'Fare']].head())
print("\nNote: Min-Max Scaling scales values between 0 and 1.")
print("Standardization scales values to mean=0 and std=1.")

# Exercise 5: Compare Label Encoding vs One-Hot Encoding
print("\n----- Exercise 5: Label Encoding vs One-Hot Encoding -----")
print("""
Label Encoding:
- Converts categories into numbers (e.g., male=1, female=0)
- Good for binary categories or ordinal data
- Does NOT increase number of columns
- Used here for: Sex

One-Hot Encoding:
- Creates separate binary columns for each category
- Avoids false order relationship between categories
- Increases number of columns
- Used here for: Embarked (C, Q, S → Embarked_Q, Embarked_S)

When to use which?
- Label Encoding → Binary or Ordinal categories
- One-Hot Encoding → Nominal categories with no order
""")

print("\n✅ Lab Session 3 Completed Successfully!")
```
